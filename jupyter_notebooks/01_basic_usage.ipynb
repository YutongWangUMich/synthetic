# Elliptic curve decision boundary

See https://en.wikipedia.org/wiki/Elliptic_curve

```python
fig, axs = plt.subplots(2, 2)
a = [-2,-1]
b = [1,2]
for i in range(2):
    for j in range(2):
        ax = axs[i,j]
        X_train, y_train = datasets.make("elliptic_curve",n_samples= 500, a= a[i], b=b[j])
        ax.scatter(X_train[:,0], X_train[:,1], c=y_train)
        ax.set_title("a="+str(a[i])+ ", b="+str(b[j]))


plt.show()
```
